```python
# Imports
import requests
from bs4 import BeautifulSoup
from gensim.summarization import summarize
import pandas as pd
```

# CNN News

```python
#Declaring the url
url_cnn = 'https://www.cnnphilippines.com/news/2020/5/12/Metro-Manila-Laguna-Cebu-ECQ-COVID-19.html'
page_cnn = requests.get(url_cnn).text
```

```python
# Turn page into BeautifulSoup object to access HTML tags
soup_cnn = BeautifulSoup(page_cnn)
```

```python
# Get headline
headline_cnn = soup_cnn.find('h1').get_text()
print(headline_cnn)
```

```python
# Get text from all <p> tags
p_tags_cnn = soup_cnn.find_all('p')

# Get the text from each of the "p" tags and strip surrounding whitespace
p_tags_text_cnn = [tag.get_text().strip() for tag in p_tags_cnn]
```

```python
#Putting texts into a data frame
cnn_df = pd.DataFrame(p_tags_text_cnn)
cnn_df
```

```python
#Dropping of unnecessary rows for cleaner dataframe
cnn = cnn_df.drop(cnn_df.index[[0,1,2,3,33,34,35,36]])

#Renaming of columns
cnn.columns = ['cnn']

#Viewing of changes
cnn
```

```python
#Saving to excel 
cnn.to_excel("cnn.xlsx")
```
